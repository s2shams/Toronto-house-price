# Modifying the Size column
Since I required all the data to be numerical for data processing, I had to modify the size column and also remove any null data. My first step was to remove the sqft at the end of every size value, this was done with the following code:

```python
def remove_sqft(size):
    edited = size[0:(len(size) - 5)]
    return edited

df_copy['Size'] = df_copy.apply(lambda x: remove_sqft(x.Size), axis=1)
```

```python
df_copy
```

```python
df_copy = df_copy.dropna()
```

```python
df_copy
```

The next step was to convert the range of the value to a single numerical integer. I considered the best way to go about this was to take the average of the range as this would give the best possible estimate of the size and remane the name of the column to "Estimated Size/ft" to keep the data set accurate. It was done through the following code:

```python
def get_avg_size(size):
    if (size) == 'N/A':
        return size
    elif "-" in str(size):
        numbers = size.split("-")
        result = (int(numbers[0]) + int(numbers[1])) / 2
        return result
    elif '–' in str(size):
        numbers1 = size.split('–')
        result1 = (int(numbers1[0]) + int(numbers1[1])) / 2
        return result1
    else:
        return size
```

```python
 pd.options.mode.chained_assignment = None
```

```python
df_copy['Size'] = df_copy.apply(lambda x: get_avg_size(x.Size), axis=1)
```

```python
df_copy = df_copy.rename(columns = {'Size' : 'Estimated Size/ft'})
```
